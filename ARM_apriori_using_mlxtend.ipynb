```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

med_dataset = pd.read_csv('med_dataset.csv')
med_dataset.head()
```

```python
print(med_dataset['metformin'][0])
print(np.nan)
print(np.isnan(med_dataset['metformin'][0]))
```

```python
def encode_units(x):
    if isinstance(x, float):
        return 0
    else:
        return 1

med_data = med_dataset.applymap(encode_units)
med_data.head()
```

```python
frequent_itemsets = apriori(med_data, min_support=0.00002, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 3) &
       (rules['confidence'] >= 0.75) ]
```

```python
# rules[ (rules['lift'] >= 3) &
#        (rules['confidence'] >= 0.75) ]

# rules['antecedents'].apply(len)
# rules['consequents'].apply(len)
rules['length_a'] = 1
rules['length_c'] = 2

for index, row in rules.iterrows():
    rules.loc[index, 'length_a'] = len(row['antecedents'])
    rules.loc[index, 'length_c'] = len(row['consequents'])

rules
```
