```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
main_dataset = pd.read_csv("dataset.csv")
main_dataset.drop(main_dataset.columns[0], axis=1, inplace=True)
main_dataset
```

```python
# only extract relevant data
wanted_cols = ['race','gender','age','A1Cresult','max_glu_serum']
ppl_dataset = main_dataset[wanted_cols]
ppl_dataset[(ppl_dataset['A1Cresult'] != 'None') &
           (ppl_dataset['max_glu_serum'] != 'None')]
```

```python
white_data = ppl_dataset[(ppl_dataset['race'] == 'Caucasian') & (ppl_dataset['gender'] == 'Male') & (ppl_dataset['A1Cresult'] != 'None') &
           (ppl_dataset['max_glu_serum'] != 'None')]
white_data
```

```python
white_data['age'].value_counts(normalize=True)
ages = ['[50-60)','[70-80)','[60-70)','[80-90)','[40-50)','[30-40)','[10-20)']
for age in ages:
    white_data[age] = 12
white_data
```

```python
for index, row in white_data.iterrows():
    print('checking row',index)
    for age in ages:
        if row['age'] == age:
            white_data.loc[index, row['age']] = 1
#             white_data[row['age']] = 1
        else:
            white_data.loc[index, age] = 0
            
white_data 
```

## Dataset to be passed thru algo

```python
white_data.drop(white_data.columns[[1,2,3,4]],axis=1, inplace=True)
white_data['race'] = 1
white_data
```

```python
frequent_itemsets = apriori(white_data, min_support=0.0001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
