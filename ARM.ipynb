```python
for col in no_readmin_cleaned:
    print(col)
    for index, row in no_readmin_cleaned.iterrows():
        if row[col] == 'Steady' or row[col] == 'Up' or row[col] == 'Down':
            row[col] = col
    print('iterated through whole col')
    
print('new dataset created')
```
